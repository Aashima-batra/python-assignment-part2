```python
# -------------------- DATA --------------------

menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}


# -------------------- TASK 1: MENU --------------------

print("\n===== MENU =====")

categories = []

# collect unique categories
for item in menu:
    cat = menu[item]["category"]
    if cat not in categories:
        categories.append(cat)

# print menu category-wise
for cat in categories:
    print("\n=====", cat, "=====")
    for item in menu:
        if menu[item]["category"] == cat:
            price = menu[item]["price"]
            status = "Available" if menu[item]["available"] else "Unavailable"
            print(item, "₹" + str(price), "[" + status + "]")

# calculations
total_items = len(menu)

available_items = 0
for item in menu:
    if menu[item]["available"]:
        available_items += 1

max_item = ""
max_price = 0

for item in menu:
    if menu[item]["price"] > max_price:
        max_price = menu[item]["price"]
        max_item = item

print("\nTotal items:", total_items)
print("Available items:", available_items)
print("Most expensive:", max_item, "₹", max_price)

print("Items under 150:")
for item in menu:
    if menu[item]["price"] < 150:
        print(item, "₹", menu[item]["price"])
```

```python
# -------------------- TASK 2: CART --------------------

cart = []

def show_cart():
    if len(cart) == 0:
        print("Cart is empty")
    else:
        for c in cart:
            print(c["item"], "x", c["quantity"])


def add_item(name, qty):
    if name not in menu:
        print(name, "not found")
        return

    if not menu[name]["available"]:
        print(name, "not available")
        return

    found = False
    for c in cart:
        if c["item"] == name:
            c["quantity"] += qty
            found = True

    if not found:
        cart.append({
            "item": name,
            "quantity": qty,
            "price": menu[name]["price"]
        })

    print(name, "added")
    show_cart()


def remove_item(name):
    found = False
    for c in cart:
        if c["item"] == name:
            cart.remove(c)
            found = True
            print(name, "removed")

    if not found:
        print(name, "not in cart")

    show_cart()


# operations
print("\n--- CART OPERATIONS ---")

add_item("Paneer Tikka", 2)
add_item("Gulab Jamun", 1)
add_item("Paneer Tikka", 1)

add_item("Mystery Burger", 1)
add_item("Chicken Wings", 1)

remove_item("Gulab Jamun")


# order summary
print("\n========== Order Summary ==========")

subtotal = 0

for c in cart:
    total = c["quantity"] * c["price"]
    subtotal += total
    print(c["item"], "x", c["quantity"], "₹", total)

print("------------------------------------")

gst = subtotal * 0.05
final_total = subtotal + gst

print("Subtotal:", round(subtotal, 2))
print("GST (5%):", round(gst, 2))
print("Total Payable:", round(final_total, 2))

print("====================================")
```

```python
# -------------------- TASK 3: INVENTORY --------------------

import copy

print("\n--- INVENTORY TRACKER ---")

# deep copy
inventory_backup = copy.deepcopy(inventory)

# test change
inventory["Paneer Tikka"]["stock"] = 5

print("\nChecking deep copy:")
print("Current:", inventory["Paneer Tikka"])
print("Backup :", inventory_backup["Paneer Tikka"])

# restore original
inventory["Paneer Tikka"]["stock"] = inventory_backup["Paneer Tikka"]["stock"]

# update stock based on cart
print("\nUpdating stock...")

for c in cart:
    item = c["item"]
    qty = c["quantity"]

    if item in inventory:
        available = inventory[item]["stock"]

        if available >= qty:
            inventory[item]["stock"] = available - qty
        else:
            print("Not enough stock for", item)
            print("Only", available, "left")
            inventory[item]["stock"] = 0

# reorder alerts
print("\nReorder Alerts:")

for item in inventory:
    stock = inventory[item]["stock"]
    level = inventory[item]["reorder_level"]

    if stock <= level:
        print("Alert:", item, "- only", stock, "left (level:", level, ")")

# final print
print("\nFinal Inventory:")
for i in inventory:
    print(i, "->", inventory[i])

print("\nBackup Inventory (unchanged):")
for i in inventory_backup:
    print(i, "->", inventory_backup[i])






# -------------------- DATA --------------------

menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}
```

```python
# -------------------- TASK 4: SALES LOG ANALYSIS --------------------

print("\n--- SALES LOG ANALYSIS ---")

# ---------- total revenue per day ----------

print("\nRevenue per day:")

daily_revenue = {}

for date in sales_log:
    total = 0
    for order in sales_log[date]:
        total += order["total"]
    daily_revenue[date] = total
    print(date, "-> ₹", total)


# ---------- best selling day ----------

best_day = ""
max_revenue = 0

for date in daily_revenue:
    if daily_revenue[date] > max_revenue:
        max_revenue = daily_revenue[date]
        best_day = date

print("\nBest selling day:", best_day, "with ₹", max_revenue)


# ---------- most ordered item ----------

item_count = {}

for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            if item not in item_count:
                item_count[item] = 1
            else:
                item_count[item] += 1

most_item = ""
max_count = 0

for item in item_count:
    if item_count[item] > max_count:
        max_count = item_count[item]
        most_item = item

print("\nMost ordered item:", most_item, "-", max_count, "times")


# ---------- add new day ----------

sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0},
]

print("\nAfter adding new day:")


# reprint revenue
print("\nUpdated Revenue per day:")

daily_revenue = {}

for date in sales_log:
    total = 0
    for order in sales_log[date]:
        total += order["total"]
    daily_revenue[date] = total
    print(date, "-> ₹", total)


# best day again
best_day = ""
max_revenue = 0

for date in daily_revenue:
    if daily_revenue[date] > max_revenue:
        max_revenue = daily_revenue[date]
        best_day = date

print("\nUpdated Best Day:", best_day, "₹", max_revenue)


# ---------- enumerate all orders ----------

print("\nAll Orders:")

count = 1

for date in sales_log:
    for order in sales_log[date]:
        print(str(count) + ".", "[" + date + "]",
              "Order #" + str(order["order_id"]),
              "— ₹" + str(order["total"]),
              "— Items:", ", ".join(order["items"]))
        count += 1
```
